## **Text-to-speech**

### **Japanese**

```python
import IPython.display as ipd
from tts import synthesize
```

```python
text = "これこそが知恵の殿堂。私のおうちにいっらしゃい。知識をあなたにも。"

audio = synthesize(text, lang="ja")
```

```python
ipd.display(ipd.Audio(audio, rate=22050, normalize=False))
```
